DATA PRE-PROCESSING

CONVERTING STRING INTO BYTES USING DICT ENCODING

```python
gender_dict = {'F':0, 'M':1}
df_train['Gender'] = df_train['Gender'].apply(lambda x: gender_dict[x])
df_train.head()
```

```python
test_dict = {'F':0, 'M':1}
df_test['Gender'] = df_test['Gender'].apply(lambda x: test_dict[x])
df_test.head()
```

DATA NORMALIZATION FOR FITTING PURPOSE

```python
cols = ['Age', 'City_Category', 'Stay_In_Current_City_Years']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in cols:
    df_train[col] = le.fit_transform(df_train[col])
df_train.tail()
```

```python
test_cols = ['Age', 'City_Category', 'Stay_In_Current_City_Years']
le = LabelEncoder()
for col in cols:
    df_test[col] = le.fit_transform(df_test[col])
df_test.tail()
```

DROPPING THE PURCHASE COLUMN FROM THE TRAIN SET FOR FITTING THE LINEAR REGRESSION MODEL

```python
X = df_train.drop(columns=['User_ID', 'Product_ID', 'Purchase'])
y = df_train['Purchase']
```

```python
test=df_test.drop(columns=['User_ID', 'Product_ID'])
```

MODEL TRANING

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn import linear_model
regr = linear_model.LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
regr.fit(x_train, y_train)
```
